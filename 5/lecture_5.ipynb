# Matplotlib

Python-скрипты можно запускать в разных оболочках. От этого зависит работа пакета Matplolib:
1. Сценарий (файл с кодом)
2. Командная оболочка (интерпретатор, например IPython)
3. Jupyter Notebook

## plt.show
1. Сценарий Python (в т.ч. в IDE).

    plt.show - данную команду запускают тогда, когда надо отобразить все созданные до этого фигуры(запускается только один раз)
    
   Основа изображений - объекты класса Figure


```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

№ 2. Командная оболочка IPython

%matplolib

plt.draw() (опционально в случае проблем, для вывода графика)

№ 3.Jupyter

P.S. В PyCharm при работе с Notebook без использования % команд, график в неинтерактивном режиме будет отображаться в любом случае, если в ячейке есть plt.show() (даже до самой команды plot)

## Сохранение изображений в файл (plt.savefig)

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.savefig('saved_image.png')

plt.show()
```

Поддерживаемые форматы изображений для сохранения:

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()

fig.savefig('saved_image1.png')

print(fig.canvas.get_supported_filetypes())
```

# Два способа вывода графиков

1. MATLAB - подобный стиль (функции)
2. Объектно-ориентированный стиль (методы объектов)

```python
# 1. MATLAB
plt.figure()

plt.subplot(2, 1, 1) # Создание сетки для двух графиков (вертикально один над другим) и выбор первого элемента 
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2) # Выбор второго элемента 
plt.plot(x, np.cos(x))

plt.show()
```

plt.figure - создание объект-контейнера, который в свою очередь содержит другие граф. объекты (систему координат (axes), тексты, метки и т.д.)
 
axes - объект осей (системы координат, в которой рисуется график), содержит графики (линии, точки и т.п.), деления, метки и т.п.

```python
# 2. ОО-подход
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))

plt.show()
```

## Декорация графиков

### 1. Цвета линий color : 
- 'blue' (строковая константа)
- 'rgbcmyk' -> 'rg' (короткие коды-буквы цветов)
- '0.14' (градация серого от 0 до 1)
- 'RRGGBB' - 'FF00EE' (16-ричный формат RGB)
- (1.0, 0.2, 0.3) (RGB)
- 'salmon' (HTML)
### 2. Стиль линий (linestyle): 
- сплошная '-', 'solid'
- штриховая '--', 'dashed'
- штрих-пунктирная '-.', 'dashdot'
- пунктирная ':', 'dotted'

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x), color = 'blue', linestyle='solid')
ax.plot(x, np.sin(x - 1), color = 'g', linestyle='dashed')
ax.plot(x, np.sin(x - 2), color = '0.75', linestyle='dashdot')
ax.plot(x, np.sin(x - 3), color = '#FF00EE', linestyle='dotted')
ax.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3))
ax.plot(x, np.sin(x - 5), color = 'salmon')
ax.plot(x, np.sin(x - 6), '--k')

plt.show()
```

## Оси (axes) (Установка пределов)

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(4)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))
ax[2].plot(x, np.sin(x))
ax[3].plot(x, np.sin(x))

ax[1].set_xlim(-2, 12)
ax[1].set_ylim(-1.5, 1.5)

ax[2].set_xlim(12, -2)
ax[2].set_ylim(1.5, -1.5)

ax[3].autoscale(tight=True)

plt.show()
```

## Метки осей, заголовок и легенда

```python
x = np.linspace(0, 10, 100)

plt.subplot(3, 1, 1)
plt.plot(x, np.sin(x))

plt.title("Синус")
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.subplot(3, 1, 2)
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')

plt.title("Синус и косинус")
plt.xlabel('x')

plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')

plt.title("Синус и косинус")
plt.xlabel('x')
plt.axis('equal')

plt.legend()
plt.subplots_adjust(hspace = 0.5)

plt.show()
```

## Scatter (точечный график)
Для каждой точки можно управлять её параметрами

```python
x = np.linspace(0, 10, 30)

plt.scatter(x, np.sin(x), marker='o')

plt.show()
```

```python
x = np.linspace(0, 10, 30)

rng = np.random.default_rng(0)

color = rng.random(30)
sizes = 30 * rng.random(30)

plt.scatter(x, np.sin(x), marker='o', c=color, s=sizes)
plt.colorbar()

plt.show()
```

Если точек больше 1000, то функция plot предпочтительнее из-за производительности

## Визуализация погрешностей (errorbar)

```python
x = np.linspace(0, 10, 50)

dy = 0.4
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')

plt.show()
```

```python
x = np.linspace(0, 10, 50)

dy = 0.4
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
plt.fill_between(x, y-dy, y+dy, color='red')
plt.show()
```

## Трехмерный график (contour - линии уровня)

```python
def f(x,y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)

Z = f(X,Y)

plt.contour(X,Y,Z, colors='green')

plt.show()
```

cmap - карта цветов для contour

```python
def f(x,y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)

Z = f(X,Y)

plt.contour(X,Y,Z, cmap='RdGy')

plt.show()
```

## Трехмерный график с заливкой (contourf)

```python
def f(x,y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)

Z = f(X,Y)

plt.contourf(X,Y,Z, cmap='RdGy')
plt.colorbar()

plt.show()
```

Сглаживание областей цветов colorbar (imshow)

```python
def f(x,y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)

Z = f(X,Y)

plt.imshow(Z, extent=[0, 5, 0, 5], cmap='RdGy', interpolation='gaussian', origin='lower', aspect='equal')
plt.colorbar()

plt.show()
```

Совмещение сглаженного графика и линий уровня

```python
def f(x,y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)

Z = f(X,Y)

c = plt.contour(X,Y,Z, colors='red')
plt.clabel(c)

plt.imshow(Z, extent=[0, 5, 0, 5], cmap='RdGy', interpolation='gaussian', origin='lower', aspect='equal')
plt.colorbar()

plt.show()
```
